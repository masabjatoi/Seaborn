```python
import seaborn as sns
import matplotlib.pyplot as plt

flights = sns.load_dataset("flights")
flights_pivot = flights.pivot(columns = "year", index= "month", values = "passengers")
plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot, annot=True, fmt=".0f", cmap="YlGnBu", annot_kws={"size": 10})
plt.title("Average Number of Passengers per Month (by Year)")
plt.show()
plt.savefig("heatmap.pdf")
```

```python
insu = pd.read_csv("insurance_data - insurance_data.csv")
```

```python
insu
```

```python
agequan = insu["age"].quantile(0.7)
```

```python
filterage = insu[insu["age"] < agequan]
filter_data = filterage["bmi"].mean()
filter = filterage[filterage["bmi"] > filter_data]
```

```python
plt.title("Diabetic patients graph")
sns.scatterplot(data = filter , x = "age" , y = "bmi" , hue= "diabetic"  , size = "smoker" , style = "gender" )
plt.legend(bbox_to_anchor =(1,1))
```
