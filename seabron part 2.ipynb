```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings("ignore")
import plotly.express as px
```

```python
tips = sns.load_dataset("tips")
```

```python
iris = sns.load_dataset("iris")
```

```python
#axis level fucntion
sns.stripplot(data=tips, x="day", y="total_bill", palette="Set2" )
```

```python
# cat plot figure level function

sns.catplot(data = tips , x = "day" , y = "total_bill" , kind = "strip" , palette = "Set2" , jitter= 0.2 , hue = "sex" , height = 4 , aspect = 1)
```

```python
sns.catplot(data = tips, x = "day" , y = "total_bill" , kind = "swarm" , palette = "Set2" , height = 4 , aspect = 2)
plt.savefig("swarm")
```

```python
sns.swarmplot(data = tips, x = "day" , y = "total_bill"  , palette = "Set2" )
```

```python
sns.boxplot(data= tips, x = "day" , y = "total_bill" , hue = "sex" , showmeans = True , meanprops = {"marker" :"o" , "markerfacecolor" : "white" , "markeredgecolor": "black"})
plt.savefig("boxplot")
```

```python
sns.catplot(kind="box" , data= tips, x = "day" , y = "total_bill" , hue = "sex" , showmeans = True , meanprops = {"marker" :"o" , "markerfacecolor" : "white" , "markeredgecolor": "black"})
```

```python
sns.violinplot(data = tips , x = "day" ,y = "total_bill"  )
```

```python
sns.catplot(kind = "violin" ,data = tips , x = "day" ,y = "total_bill"  )
```

```python
sns.barplot(data = tips , x = "sex" , y = "total_bill" , hue = "smoker" , estimator = np.min)
```

```python

sns.catplot(data = tips , x = "sex" , y = "total_bill" , hue = "smoker" , kind= "bar")

```

```python

sns.barplot(  data = tips , x = "sex" , y = "total_bill" , hue=  "smoker"  , estimator = np.min , palette = "Set2")
```

```python
sns.pointplot(  data = tips , x = "sex" , y = "total_bill" , hue=  "smoker"  , estimator = 'mean')
```

```python
sns.countplot(data= tips , x = "sex" , hue = "day" )
```

```python

sns.catplot(data= tips , x = "sex" , y = "total_bill"  , kind = "box" , col = "smoker" , height = 4  , aspect = 1 , row = "time")
plt.show()
plt.savefig("catplot")
```

```python
sns.regplot(data = tips , x = "total_bill" , y = "tip" )
```

```python
sns.lmplot(data = tips , x = "total_bill" , y = "tip" , hue=  "sex" )

plt.savefig("Linear.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
sns.residplot(data = tips,  x = 'tip' , y = "total_bill" )
```

```python
g = sns.FacetGrid(data = tips, col= "day" , row = "time" , hue = "smoker")
g.map(sns.scatterplot , "sex" , "total_bill")
plt.legend()
```

```python
sns.pairplot(iris)
plt.savefig("pairplot")
```

```python
sns.pairplot(iris , hue = "species")
```

```python
g = sns.PairGrid(data = iris , hue = "species")
g.map_diag(sns.boxplot)
g.map_offdiag(sns.kdeplot)
```

```python
g = sns.PairGrid(data = iris , hue = "species" , vars = ["sepal_length" , "sepal_width"])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
sns.jointplot(data = tips , x= "total_bill" , y = "tip" ,kind = "reg" )
plt.savefig("jointplot")
```

```python
g = sns.JointGrid(data = tips, x = "total_bill" , y = "tip" )
g.plot(sns.kdeplot , sns.violinplot)
plt.show()
```
