```python
gapminder = px.data.gapminder()
gap = gapminder[gapminder["continent"] == "Europe"].pivot(index = "country" , columns = "year" , values = "lifeExp" )
plt.figure(figsize=(20, 20))
sns.heatmap(gap, annot=True, fmt=".1f", cmap= "summer" , linewidth = 0.5 )
plt.show()
```

```python
mask = gap.isnull()
plt.figure(figsize=(15, 15))
sns.heatmap(gap, annot=True, cmap="coolwarm", mask=mask, cbar=False)
plt.title("European Life Expectancy Over the Years (with Missing Data Highlighted)")
plt.show()
```
